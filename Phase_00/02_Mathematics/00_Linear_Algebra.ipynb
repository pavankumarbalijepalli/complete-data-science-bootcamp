__`Linear algebra is basically understanding how a vector which has a magnitude and direction will behave when it faces another vector.`__

```python
import matplotlib.pyplot as plt

# Do not modify the following function
def make_resultant(vector_1: list, vector_2: list):
    origin = np.array([0, 0])
    vector_1 = np.array(vector_1)
    vector_2 = np.array(vector_2)
    resultant = vector_2 + vector_1
    path_1 = np.array([vector_2[0] + vector_1[0], vector_2[1]])
    
    plt.quiver([0, 0, 0, vector_2[0], vector_1[0]], [0, 0, 0, vector_2[1], vector_1[1]], [vector_1[0], vector_2[0], resultant[0], vector_1[0], vector_2[0]], [vector_1[1], vector_2[1], resultant[1], vector_1[1], vector_2[1]], color=['b','r','g','c', 'y'], angles='xy', scale_units='xy', scale=1)
    plt.axhline(y=resultant[-1], linewidth=0.5, color = 'k', linestyle='-.')
    plt.axvline(x=resultant[0], linewidth=0.5, color = 'k', linestyle='-.')
    plt.axhline(y=0, linewidth=0.3, color = 'k', linestyle='-.')
    plt.axvline(x=0, linewidth=0.3, color = 'k', linestyle='-.')
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.title("Vectors, their resultant, and their path to resultant.")
    plt.show()
```

```python
# Give your random 2-dimensional vectors, to see their resultant vectors.
# Keep the values between -10 to 10
make_resultant([-5, -2], [2, 5])
```

---
__Understanding the plot:__
* In the above plot, we can observe the vectors *`Red`* and *`Blue`* will lead to *`Green`* vector.
* This *Green* Vector is called the __Resultant__ vector.
* We will obtain the *`Green`* vector, when we place the origin of *`Red`* vector on the ending of *`Blue`* vector or origin of the *`Blue`* vector on the ending of *`Red`* vector.
* That is what is seen in the *`Cyan`* & *`Yellow`* vector paths.
* You can also see that the arrows *`Cyan`* and *`Blue`* are parallel because of their origin alone is shifted.
* You can also see that the arrows *`Red`* and *`Yellow`* are parallel because of their origin alone is shifted.
